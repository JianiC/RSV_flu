```python
import re
import pandas as pd
```

```python
mean_tmp=pd.read_csv("climate-data/mean-temp-US-year-states.csv")
```

```python
mean=mean_tmp.mean(axis=1)
mean_tmp=pd.concat([mean_tmp, mean], axis=1)
```

```python
mean_tmp[['year','state']] = mean_tmp['Date-state'].str.split('-',expand=True)
mean_tmp
```

```python
dic=pd.read_csv("climate-data/state_region_dic.csv",index_col=0, squeeze=True).to_dict()
mean_tmp['HHS_region']=mean_tmp['state'].map(dic)
mean_tmp

```
